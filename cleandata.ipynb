```python
import pandas as pd
import numpy as np

# ---------- 1. Load raw datasets ----------
cases = pd.read_csv("cases.csv")
methods = pd.read_csv("methods.csv")
organisations = pd.read_csv("organisations.csv")
```

```python
# ---------- 2. Clean blank strings ----------
def clean_blanks(df):
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype(str).str.strip()
            df[col] = df[col].replace(
                ["", "None", "NULL", "null", "nan"], np.nan
            )
    return df

cases = clean_blanks(cases)
methods = clean_blanks(methods)
organisations = clean_blanks(organisations)
```

```python
# ---------- 3. Clean CASES ----------
cases["primary_org_id"] = pd.to_numeric(
    cases.get("primary_organizer"), errors="coerce"
)
```

```python
cases["country_best"] = cases["country"].fillna("Unknown")

cases["city_best"] = cases["city"]
cases.loc[cases["city_best"].isna(), "city_best"] = cases.loc[
    cases["city_best"].isna(), "location_name"
]
cases["city_best"] = cases["city_best"].fillna("Unknown")
```

```python
print(
    cases[["country", "country_best", "city", "location_name", "city_best"]].head(10)
)
```

```python
print(cases["country_best"].value_counts().head(10))
print(cases["city_best"].value_counts().head(10))
```

```python
# Data quality flags
def missing_methods_flag(x):
    if pd.isna(x):
        return 1
    x = str(x).strip()
    return 1 if x == "{}" else 0

cases["missing_methods"] = cases["specific_methods_tools_techniques"].apply(missing_methods_flag)
cases["missing_org"] = cases["primary_org_id"].isna().astype(int)
cases["missing_country"] = cases["country"].isna().astype(int)

fact_cases_clean = cases[
    [
        "id",
        "type",
        "post_date",
        "updated_date",
        "location_name",
        "city",
        "province",
        "country",
        "country_best",
        "city_best",
        "approaches",
        "method_types",
        "tools_techniques_types",
        "specific_methods_tools_techniques",
        "primary_org_id",
        "missing_org",
        "missing_methods",
        "missing_country",
    ]
].dropna(subset=["id"]).drop_duplicates(subset=["id"])
```

```python
# ---------- 4. Clean METHODS ----------
methods["method_id"] = pd.to_numeric(methods["id"], errors="coerce")

dim_methods_clean = methods[
    [
        "method_id",
        "type",
        "original_language",
        "post_date",
        "updated_date",
        "city",
        "province",
        "country",
        "method_types",
        "purpose_method",
        "completeness",
        "collections",
    ]
].dropna(subset=["method_id"]).drop_duplicates(subset=["method_id"])
```

```python
print(methods["method_id"].isna().value_counts())
```

```python
# ---------- 5. Clean ORGANISATIONS ----------
organisations["org_id"] = pd.to_numeric(organisations["id"], errors="coerce")

dim_organisations_clean = organisations[
    [
        "org_id",
        "type",
        "title",
        "creator_id",
        "creator_name",
        "city",
        "province",
        "country",
        "sector",
        "scope_of_influence_1",
        "scope_of_influence_2",
        "scope_of_influence_3",
        "collections",
    ]
].dropna(subset=["org_id"]).drop_duplicates(subset=["org_id"])
```

```python
print(organisations["org_id"].isna().value_counts())
```

```python
# ---------- 6. Export cleaned datasets ----------
fact_cases_clean.to_csv("fact_cases_clean.csv", index=False)
dim_methods_clean.to_csv("dim_methods_clean.csv", index=False)
dim_organisations_clean.to_csv("dim_organisations_clean.csv", index=False)

print("Cleaning completed successfully")
print("Cases:", len(fact_cases_clean))
print("Methods:", len(dim_methods_clean))
print("Organisations:", len(dim_organisations_clean))
```
